```python
import joblib
import pandas as pd
#load your model for further usage
gs = joblib.load("tree.pkl")
```

```python
gs.best_params_["classifier"]
```

```python
results = pd.DataFrame(gs.cv_results_)
```

```python
results.sort_values('rank_test_score')
```

```python
sorted_results = results.sort_values('rank_test_score')
```

```python
import numpy as np
from sklearn.metrics import *
with open('data.npy', 'rb') as f:
    train_data = np.load(f)
    test_data = np.load(f)
    train_label = np.load(f)
    test_label = np.load(f)
```

```python
sorted_results.iloc[100]
```

```python
# e.g. Accuracy, Precision etc.
# PERFORMANCE ON TRAINING DATASET
#best_clf = a.best_params_["classifier"]
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import *

b_result = sorted_results.iloc[0]
best_clf = DecisionTreeClassifier(ccp_alpha = b_result["param_classifier__ccp_alpha"],
                                  criterion = b_result["param_classifier__criterion"],
                                  max_depth = b_result["param_classifier__max_depth"],
                                  max_leaf_nodes = b_result["param_classifier__max_leaf_nodes"],
                                  min_impurity_decrease = b_result["param_classifier__ccp_alpha"],
                                  min_samples_leaf = b_result["param_classifier__min_samples_leaf"],
                                  min_samples_split = b_result["param_classifier__min_samples_split"],
                                  splitter = b_result["param_classifier__splitter"])
best_clf.fit(train_data,train_label)
prediction = best_clf.predict(train_data)

accuracy_tf = accuracy_score(train_label,prediction)
precision_tf = precision_score(train_label,prediction)
recall_tf = recall_score(train_label,prediction)
f1_tf = f1_score(train_label,prediction)
confusion_matrix_tf = confusion_matrix(train_label, prediction)

print("Accuracy with transformed data: ",accuracy_tf)
print("Precision with transformed data: ",precision_tf)
print("Recall with transformed data: ",recall_tf)
print("F1 score with transformed data: ",f1_tf)

print(confusion_matrix_tf)
prediction = best_clf.predict(test_data)

accuracy_tf = accuracy_score(test_label,prediction)
precision_tf = precision_score(test_label,prediction)
recall_tf = recall_score(test_label,prediction)
f1_tf = f1_score(test_label,prediction)
confusion_matrix_tf = confusion_matrix(test_label, prediction)

print("Accuracy with transformed data: ",accuracy_tf)
print("Precision with transformed data: ",precision_tf)
print("Recall with transformed data: ",recall_tf)
print("F1 score with transformed data: ",f1_tf)
print(confusion_matrix_tf)
```
